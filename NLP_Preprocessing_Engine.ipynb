# Task 1 — Conceptual Understanding

## 1. Difference between "Love" and "love" in NLP

In NLP, "Love" and "love" are treated as different words unless text normalization is applied. Converting text to lowercase helps reduce duplication and improves consistency.

## 2. What happens if stopwords are not removed?

If stopwords are not removed:
- Model size increases
- Processing becomes slower
- Important keywords may get less importance

## 3. Two scenarios where removing stopwords is harmful

1. Sentiment Analysis:
Removing words like "not" changes meaning.

Example:
"not good" becomes "good"

2. Question Answering Systems:
Words like "how", "when", and "where" are important.

## 4. Difference between stemming and lemmatization

Stemming removes word endings using simple rules.

Example:
"playing" becomes "play"

Lemmatization converts words to meaningful dictionary form.

Example:
"better" becomes "good"

```python
def preprocess_text(text):
    if not isinstance(text, str):
        return [], ""

    
    text = text.lower()
    text = re.sub(r'http\\S+|www\\S+', '', text)
    text = re.sub(r'\\S+@\\S+', '', text)
    text = re.sub(r'\\d+', '', text)
    text = re.sub(r'[^a-zA-Z\\s]', '', text)
    text = re.sub(r'(\\w)\\1{2,}', r'\\1', text)
    text = re.sub(r'\\s+', ' ', text).strip()
    tokens = text.split()
    cleaned_tokens = []

    for token in tokens:

        if len(token) > 2 or token in ["no", "not"]:
            cleaned_tokens.append(token)

    cleaned_sentence = " ".join(cleaned_tokens)

    return cleaned_tokens, cleaned_sentence
```

```python
sentences = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]
```

```python
all_tokens = []

for sentence in sentences:

    tokens, cleaned = preprocess_text(sentence)

    all_tokens.extend(tokens)

    print("Original Text:")
    print(sentence)

    print("Cleaned Tokens:")
    print(tokens)

    print("Cleaned Sentence:")
    print(cleaned)

    print("-" * 50)
```

```python
print("TOKEN ANALYTICS\\n")

for sentence in sentences:

    tokens, cleaned = preprocess_text(sentence)

    total_tokens = len(tokens)

    unique_tokens = len(set(tokens))

    avg_length = 0

    if total_tokens > 0:
        avg_length = sum(len(t) for t in tokens) / total_tokens

    print(f"Sentence: {sentence}")
    print(f"Total Tokens: {total_tokens}")
    print(f"Unique Tokens: {unique_tokens}")
    print(f"Average Token Length: {avg_length:.2f}")

    print("-" * 50)
```
